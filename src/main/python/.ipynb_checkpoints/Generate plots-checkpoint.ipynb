```python
project = None
output = None
pixel_to_micrometers = 0.016
```

```python
import z5py
from pathlib import Path
import os
import matplotlib.pyplot as plt
# local imports
from plot import subplot_histogram, subplot_histogram_compare, read_table, subplot_violin_compare, subplot_violin
from cell import Cell
```

```python
def plot_histogram(cell, table, column, minx, maxx, maxy, title, prefix):
    fig, ax = plt.subplots(figsize=(3, 3))
    subplot_histogram(cell, table, column, minx, maxx, maxy, ax, pixel_to_micrometers)
    ax.set_xlabel(xlabel)
    render_and_save(fig, title, prefix)


def plot_histogram_compare(cell, table, column, compare_column, minx, maxx, maxy, labels, colors, title, xlabel, prefix):
    fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(10, 3))

    subplot_histogram(cell, table, column, min_x, max_x, max_y, ax0, pixel_to_micrometers)
    ax0.set_xlabel(xlabel)

    subplot_histogram_compare(table=read_table(cell, table), column=column,
                              compare_column=compare_column, min_x=minx, max_x=maxx, ax=ax1,
                              text_true=labels[1], text_false=labels[0],
                              color_true=colors[1], color_false=colors[0])
    ax1.set_xlabel(xlabel)

    subplot_violin_compare(cell, table, column=column, compare_column=mt_connected_column, ax=ax2, 
                           text_true=text_connected, text_false=text_not_connected, 
                           color_true=color_connected, color_false=color_not_connected)
    ax2.set_ylabel(xlabel)

    render_and_save(fig, title, prefix)

    
def plot_histogram_violin(cell, table, column, minx, maxx, maxy, title, prefix):
    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(7, 3))

    subplot_histogram(cell, table, column, min_x, max_x, max_y, ax0, pixel_to_micrometers)
    ax0.set_xlabel(xlabel)

    subplot_violin(cell, table, column=column, ax=ax1)
    ax1.set_xlabel(xlabel)

    render_and_save(fig, title, prefix)

    
def render_and_save(fig, title, prefix):
    fig.suptitle(title)
    plt.tight_layout()
    if output:
        plt.savefig(str(Path(output).joinpath(prefix + "_" + title.replace(" ", "_") + ".svg")))
    plt.show()
    plt.close(fig)
```

## Input data

```python
project_name = Path(project).name.rstrip('.n5')
cell = Cell(project_name, Path(project), '#edb380')
if output:
    Path(output).mkdir(parents=True, exist_ok=True)
    print("Saving plots to %s" % output)
```

## Secretory granules

```python
table = 'granules_individual.csv'
mt_connected_column = 'connected to microtubule'
text_connected = 'assoc'
text_not_connected = 'not assoc'
color_connected = '#ff7254'
color_not_connected = '#616163'
prefix = "all_granules"
```

```python
# Plot distance to MT statistics

title = 'Secretory granule distance to microtubules'
xlabel = 'Volume in μm³'
column = 'distance to microtubule in um'
min_x = 0; max_x = 1; max_y = 0.5

plot_histogram(cell, table, column, min_x, max_x, max_y, title, prefix)
```

```python
# Plot volume

title = 'Secretory granule volume'
xlabel = 'Volume in μm³'
column = 'size in um^3'
min_x = 0; max_x = 0.1; max_y = 0.2

plot_histogram_compare(cell, table, column, mt_connected_column, min_x, max_x, max_y, 
                       labels=[text_not_connected, text_connected], colors=[color_not_connected, color_connected],
                        title=title, xlabel=xlabel, prefix=prefix)
```

```python
# Plot distance to membrane

title = 'Secretory granule distance to membrane'
xlabel = 'Distance to membrane in μm'
column = 'distance to membrane in um'

min_x = 0; max_x = 4; max_y = 0.35

plot_histogram_compare(cell, table, column, mt_connected_column, min_x, max_x, max_y, 
                       labels=[text_not_connected, text_connected], colors=[color_not_connected, color_connected],
                        title=title, xlabel=xlabel, prefix=prefix)
```

```python
# Plot distance to nucleus

title = 'Secretory granule distance to nucleus'
xlabel = 'Distance to nucleus in μm'
column = 'distance to nucleus in um'
min_x = 0; max_x = 12; max_y = 0.13

plot_histogram_compare(cell, table, column, mt_connected_column, min_x, max_x, max_y, 
                       labels=[text_not_connected, text_connected], colors=[color_not_connected, color_connected],
                        title=title, xlabel=xlabel, prefix=prefix)
```

```python
# Plot distance to golgi

title = 'Secretory granule distance to golgi'
xlabel = 'Distance to golgi in μm'
column = 'distance to golgi in um'
min_x = 0; max_x = 5; max_y = 0.6

plot_histogram_compare(cell, table, column, mt_connected_column, min_x, max_x, max_y, 
                       labels=[text_not_connected, text_connected], colors=[color_not_connected, color_connected],
                        title=title, xlabel=xlabel, prefix=prefix)
```
